```python
import pandas as pd

titanic_tr = pd.read_csv("titanic_train.csv")  # cargamos la base de datos de Titanic

titanic_tr.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

df = titanic_tr.dropna(axis=0)  # eliminamos las filas con valores nulos

age = df[['Age']]  # seleccionamos solamente la edad
survived = df['Survived']  # supervivientes

clf = RandomForestClassifier()
clf.fit(age, survived)  # conjunto de valores de entrenamiento. age: valores de entrada, survived: valores de salida

clf.predict(age)

cross_val_score(clf, age, survived)  # cross-validation para evaluar el modelo
```

```python
df_2 = df[['Age', 'Pclass', 'Sex']]  # seleccionamos 3 features

df_2['Sex'] = pd.factorize(df_2['Sex'])[0]  # transformamos el sexo a números

clf_2 = RandomForestClassifier()
clf_2.fit(df_2, survived)

clf_2.predict(df_2)

cross_val_score(clf_2, df_2, survived)  # cross-validation
```

```python
n_estimators = 1

for i in range(10):
    clf_3 = RandomForestClassifier(n_estimators=n_estimators)
    
    clf_3.fit(df_2, survived)
    clf_3.predict(df_2)
    
    score = cross_val_score(clf_2, df_2, survived)  # cross-validation
    
    print('Número de estimadores %i' % n_estimators)
    print(score[0])
    
    n_estimators += 10
```

```python
names = titanic_tr['Name']
surnames = {}

# Creamos un diccionario aue tenga como clave los apellidos de los viajeros y valor el número de apariciones
for name in names:
    surname = name.split(',')[0]
    times = surnames.get(surname)
    
    if times:
        times += 1
        surnames.update({surname: times})
    else:
        surnames.update({surname: 1})
```

```python
times = []
for i in range(len(titanic_tr['Name'])):
    surname = titanic_tr['Name'][i].split(',')[0]
    times.append(surnames.get(surname))
    
titanic_tr['FamilyMembers'] = times  # nueva columna con el número de familiares

titanic_tr.head()
```

```python
df_3 = titanic_tr.dropna(axis=0)  # eliminamos las filas con valores nulos
df_4 = df_3[['Age', 'Pclass', 'Sex', 'FamilyMembers']]  # seleccionamos 4 features

df_4 = df_4.dropna(axis=0)  # eliminamos las filas con valores nulos

df_4['Sex'] = pd.factorize(df_4['Sex'])[0]  # transformamos el sexo a números

survived = df_3['Survived']  # supervivientes

clf_4 = RandomForestClassifier()
clf_4.fit(df_4, survived)

clf_4.predict(df_4)

cross_val_score(clf_4, df_4, survived)  # cross-validation
```
