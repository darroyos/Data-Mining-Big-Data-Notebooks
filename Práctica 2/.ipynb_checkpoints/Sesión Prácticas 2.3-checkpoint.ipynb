```python
import pandas as pd

ex = pd.read_csv("ex.csv", names = ["Name", "Sex", "Age", "Country", "Level"])

ex.to_csv("ex2.csv", index=False)
```

```python
ex2 = pd.read_csv("ex2.csv")
```

```python
ex2 = ex2.dropna(axis=0, how='any')
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(ex2['Sex'])
ex2['Sex'] = le.transform(ex2['Sex'])

ex2.head()
```

```python
ex2 = ex2[ex2['Age'] > 0]
ex2.head()
```

```python
ex2['NameLen'] = [len(name) for name in ex2['Name']]

ex2.head()
```

```python
ex2['Hemisphere'] = [1, 0, 1, 1]

ex2.head()
```

```python
sex_dummies = pd.get_dummies(ex2['Sex'], prefix='Level')

pd.concat([ex2, sex_dummies], axis=1)
```
