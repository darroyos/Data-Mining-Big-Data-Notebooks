```python
temp = []

with open("timeSerie.txt", mode="r") as timeSerie:
    for line in timeSerie:
        line = line.strip()  # eliminamos los saltos de línea
        line = line.replace(",", ".")  # reemplazamos las comas por puntos
        
        if line == "err":
            temp.append(-1)
        else:
            temp.append(float(line))
```

```python
for i in range(len(temp)):
    if temp[i] == -1:  # valor erróneo
        if i == 0:
            temp[i] = temp[i + 1]
        elif i == len(temp) - 1:
            temp[i] = temp[i - 1]
        else:
            temp[i] = (temp[i - 1] + temp[i + 1]) / 2
```

```python
print(temp)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-0.2, 1.2, num=len(temp), endpoint=False)
y1 = np.array(temp)

plt.plot(x1, y1, 'o')
```

```python
x2 = np.linspace(0, len(temp), num=len(temp), endpoint=False)
plt.plot(x2, y1)
```

```python
variacion = []

for i in range(len(temp)):
    if i == 0:
        variacion.append(0)
    else:
        variacion.append(round(temp[i] - temp[i - 1], 2))
        
print(variacion)
```

```python
x3 = np.linspace(-2, 16, num=len(variacion), endpoint=False)

plt.plot(x3, variacion, 'o-')
```
